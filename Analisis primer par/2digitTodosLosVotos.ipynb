```python
import matplotlib.pyplot as plt
import numpy as np
import math
import pandas as pd 
letters = {}
for i in range(9,99):
    letters[str(i+1)] = 0

nfib = 304
```

```python
df = pd.read_csv("https://docs.google.com/spreadsheets/d/17wrVuGxZWhWtiZ7y6Q59CGSSl5K-XR9a/export?format=csv")
df
```

```python
pl = []
for i in df['PL']:
    pl.append(i)
pl
```

```python
pn = []
for i in df['PN']:
    pn.append(i)
pn
```

```python
pli = []
for i in df['LIBRE-PINU']:
    pli.append(i)
pli
```

```python
tot = []
for i in df['TOTAL']:
    tot.append(i)
tot
```

```python
list_of_nums = pl
for n in list_of_nums:
    first_digit = str(n)[0]+str(n)[1]
    letters[first_digit] += 1
y = [round(val / nfib, 4) for val in letters.values()]
fre_relativa = y
y = list(letters.values())
f0= y
x = list(letters.keys())
y =  [round(math.log10(1+(1/int(val))), 4) for val in letters.keys()]
benford = y
```

```python
f0 = np.array(f0)
f0
```

```python
fre_relativa = np.array(fre_relativa)
fre_relativa 
```

```python
benford = np.array(benford)
benford
```

```python
f0_tot=f0.sum()
f0_tot
```

```python
fe = benford*f0_tot
```

```python
fe 
```

```python
fe.sum()
```
